
### Data Cleaning
> **Tip**: Make sure that you keep your reader informed on the steps that you are taking in your investigation. Based on the initial exploration, the gun dataset has three issues that need attention:

1. **`month` is stored as a string** — we convert it to a proper `datetime` so we can do time-based analysis.
2. **Many columns have missing values** — columns like `other`, `permit_recheck`, `rentals_*`, and `returned_*` are sparse because the FBI added those reporting categories in later years. We fill these with `0` since a missing value here means "not yet reported," not "unknown amount."
3. **Duplicates** — we check for and drop any duplicate `(state, month)` rows.

```python
# 1. Convert 'month' to datetime and extract year for later grouping
df['month'] = pd.to_datetime(df['month'])
df['year'] = df['month'].dt.year
print("month dtype:", df['month'].dtype)
df[['month','year']].head()
```

```python
# 2. Fill NaN in numeric gun-type columns with 0
#    (missing = FBI had not yet added that category for that month)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

def fill_missing_with_zero(frame, columns):
    """Fill NaN in the given columns with 0 and return how many cells were filled."""
    before = frame[columns].isnull().sum().sum()
    frame[columns] = frame[columns].fillna(0)
    return before

filled = fill_missing_with_zero(df, numeric_cols)
print(f"Filled {filled} missing numeric cells with 0")
print(f"Remaining nulls in df: {df.isnull().sum().sum()}")
```

```python
# 3. Check for duplicate state-month rows
dup_count = df.duplicated(subset=['state','month']).sum()
print(f"Duplicate (state, month) rows: {dup_count}")

# Drop if any exist
if dup_count > 0:
    df = df.drop_duplicates(subset=['state','month']).reset_index(drop=True)
    print(f"After dedup, shape: {df.shape}")
else:
    print("No duplicates found — shape unchanged:", df.shape)
```

```python
# Final sanity check on the cleaned gun dataframe
print("Shape:", df.shape)
print("Date range:", df['month'].min(), "→", df['month'].max())
print("Unique states/territories:", df['state'].nunique())
df.info()
```

### Loading the U.S. Census Data

The census dataset is structured in **wide format** — each row is a demographic fact and each column is a U.S. state. We need to transpose it so that each row represents one state, then extract the key columns we'll use for analysis (population and demographic percentages).

```python
# Load the census data (wide format: rows=facts, columns=states)
census_raw = pd.read_csv('us_census_data.csv')

# Transpose so rows = states, columns = facts; set state names as index
census_T = census_raw.set_index('Fact').drop(columns=['Fact Note']).T
census_T.index.name = 'state'
census_T.reset_index(inplace=True)

# Select only the columns we'll use for analysis
CENSUS_COLS = {
    'Population estimates, July 1, 2016,  (V2016)': 'population_2016',
    'Persons under 18 years, percent, July 1, 2016,  (V2016)': 'pct_under18',
    'Persons 65 years and over, percent,  July 1, 2016,  (V2016)': 'pct_65plus',
    'Female persons, percent,  July 1, 2016,  (V2016)': 'pct_female',
    'White alone, not Hispanic or Latino, percent, July 1, 2016,  (V2016)': 'pct_white',
    'Black or African American alone, percent, July 1, 2016,  (V2016)': 'pct_black',
    'Hispanic or Latino, percent, July 1, 2016,  (V2016)': 'pct_hispanic',
}

census = census_T[['state'] + [c for c in CENSUS_COLS if c in census_T.columns]].copy()
census.rename(columns=CENSUS_COLS, inplace=True)

# Strip commas/percent signs and convert to numeric
def clean_numeric(series):
    """Remove commas/percent signs and cast to float."""
    return pd.to_numeric(series.astype(str).str.replace(',', '', regex=False)
                                           .str.replace('%', '', regex=False)
                                           .str.strip(), errors='coerce')

for col in census.columns[1:]:
    census[col] = clean_numeric(census[col])

print(f"Census shape: {census.shape}")
census.head()
```

### Merging FBI Gun Data with Census Data

The gun dataset has one row per **state × month**, covering multiple years. Before merging with the census data (which provides a single 2016 population snapshot), we aggregate the gun data by state — summing the total background checks and gun type counts across all months. We then perform an **inner join** on state name, which drops entries not present in both datasets (e.g., "District of Columbia" and territories absent from the census table).

```python
# Aggregate gun data by state (sum numeric columns across all months)
gun_by_state = (df.groupby('state', as_index=False)
                  .agg(total_checks=('totals', 'sum'),
                       total_handgun=('handgun', 'sum'),
                       total_long_gun=('long_gun', 'sum'),
                       total_permit=('permit', 'sum'),
                       total_multiple=('multiple', 'sum')))

# Inner join — keeps only states present in both datasets
merged = gun_by_state.merge(census, on='state', how='inner')

# Derived column: background checks per 100k residents
merged['checks_per_100k'] = (merged['total_checks'] / merged['population_2016']) * 100_000

print(f"States in gun data:    {gun_by_state['state'].nunique()}")
print(f"States in census data: {census['state'].nunique()}")
print(f"States after merge:    {merged['state'].nunique()}")
print(f"\nMerged dataframe shape: {merged.shape}")
merged.head()
```

```python
# Verify no nulls in key columns after merge
print("Null counts in merged dataset:")
merged[['total_checks', 'population_2016', 'checks_per_100k']].isnull().sum()
```

### Research Question 1: How have FBI gun background checks trended over time, and which states drive the most activity?

We'll approach this question from two angles:
- **1D (time):** look at total monthly background checks aggregated nationally to see long-term trends and seasonality.
- **2D (state × total):** identify the top-contributing states and compare them using the merged gun + census data.

```python
# --- 1D analysis: national totals over time ---
# Sum checks across all states for each month
monthly_totals = df.groupby('month', as_index=False)['totals'].sum()

# Reusable helper: label and show a matplotlib plot consistently
def annotate_plot(title, xlabel, ylabel):
    """Apply title and axis labels to the current matplotlib axes."""
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()

# Line plot of total checks per month
plt.figure(figsize=(11, 4))
plt.plot(monthly_totals['month'], monthly_totals['totals'], color='steelblue')
annotate_plot('FBI Gun Background Checks per Month (U.S. Total)',
              'Month', 'Total Background Checks')
plt.show()

# Summary statistics
print(f"Mean monthly total:   {monthly_totals['totals'].mean():,.0f}")
print(f"Median monthly total: {monthly_totals['totals'].median():,.0f}")
print(f"Peak month:           {monthly_totals.loc[monthly_totals['totals'].idxmax(), 'month'].date()}"
      f" ({monthly_totals['totals'].max():,} checks)")
```

**Observation:** National background checks show a strong **upward trend** from 1998 through the mid-2010s, with pronounced **annual seasonality** — peaks near November/December (holiday season) and dips in the summer months. The spike visible near the end corresponds to late-2015 / early-2016, a well-documented period of elevated firearm purchasing activity.

```python
# --- 2D analysis: which states contribute the most to total checks? ---
# Use the merged dataframe (built in the Data Wrangling section) for state-level totals
top10 = merged.nlargest(10, 'total_checks')[['state', 'total_checks', 'population_2016', 'checks_per_100k']]
print("Top 10 states by total background checks (all years):")
print(top10.to_string(index=False))

# Bar plot — different plot type than the line chart above
plt.figure(figsize=(10, 5))
plt.bar(top10['state'], top10['total_checks'], color='darkorange')
plt.xticks(rotation=45, ha='right')
annotate_plot('Top 10 States by Total FBI Background Checks',
              'State', 'Total Checks (all years)')
plt.show()
```

**Observation:** Raw totals are dominated by populous states such as Kentucky, California, Texas, and Illinois. However, **raw totals conflate population size with per-person activity** — Kentucky leads not because of population (it is mid-sized) but because of its policy of running recurring background checks on concealed-carry permit holders.

To get a fairer comparison we should look at **checks per 100,000 residents**, which the next cell explores.

```python
# Normalize by population: top 10 by checks_per_100k
top10_rate = merged.nlargest(10, 'checks_per_100k')[['state', 'total_checks', 'population_2016', 'checks_per_100k']]
print("Top 10 states by background checks per 100,000 residents:")
print(top10_rate.to_string(index=False))

# Correlation between population and raw total checks
corr_pop_checks = merged['population_2016'].corr(merged['total_checks'])
print(f"\nPearson correlation (population vs total_checks): {corr_pop_checks:.3f}")

# Bar plot of the per-capita leaders
plt.figure(figsize=(10, 5))
plt.bar(top10_rate['state'], top10_rate['checks_per_100k'], color='seagreen')
plt.xticks(rotation=45, ha='right')
annotate_plot('Top 10 States by Background Checks per 100k Residents',
              'State', 'Checks per 100,000 Residents')
plt.show()
```

**Findings for Research Question 1:**

1. **Time trend (1D):** National background checks have risen substantially over the 1998–2017 window, with clear end-of-year seasonal peaks.
2. **Raw state totals (2D):** Population heavyweights lead the absolute counts, and population is positively correlated with total checks (see the Pearson coefficient printed above).
3. **Per-capita picture (2D, normalized):** After normalizing by population, the ranking shifts dramatically — smaller states with specific permit-renewal policies (e.g., Kentucky) top the per-capita list. This shows why a population-adjusted metric is essential when comparing states.

*Note: correlation does not imply causation. The relationship between population and checks is mechanical (more people → more transactions) and does not by itself say anything about per-person behavior.*

### Research Question 2: Do state demographics relate to per-capita gun background check rates?

Research Question 1 looked at *how much* and *where*. Research Question 2 asks *why some states are higher than others* by comparing the dependent variable **`checks_per_100k`** against three demographic independent variables from the census:

- `pct_65plus` — percent of population aged 65 and over
- `pct_under18` — percent of population under 18
- `pct_white` — percent of non-Hispanic white residents

We'll start with a **1D distribution** of per-capita checks, then do **2D scatter plots** against each demographic variable, and finish with a **correlation matrix** to summarize the relationships.

```python
# --- 1D analysis: distribution of per-capita background checks across states ---
plt.figure(figsize=(9, 4))
plt.hist(merged['checks_per_100k'], bins=15, color='mediumpurple', edgecolor='black')
annotate_plot('Distribution of Background Checks per 100k Residents (by State)',
              'Checks per 100,000 Residents', 'Number of States')
plt.show()

# Summary statistics for the dependent variable
print(merged['checks_per_100k'].describe().round(1))
print(f"\nSkewness: {merged['checks_per_100k'].skew():.2f} (positive = long right tail)")
```

**Observation:** The distribution is **right-skewed** — most states cluster in the 50,000–150,000 per-100k range, with a small number of high-outlier states (Kentucky being the most extreme). When interpreting scatter plots below, remember the outliers can dominate the visual story.

```python
# --- 2D analysis: scatter plots of each demographic vs per-capita checks ---
# Helper function to draw a consistent scatter plot for any independent variable
def scatter_vs_checks(x_col, x_label, color):
    """Scatter plot of the given census column against checks_per_100k, with Pearson r."""
    r = merged[x_col].corr(merged['checks_per_100k'])
    plt.figure(figsize=(7, 4))
    plt.scatter(merged[x_col], merged['checks_per_100k'], color=color, alpha=0.7)
    annotate_plot(f'{x_label} vs. Checks per 100k  (r = {r:.2f})',
                  x_label, 'Checks per 100k')
    plt.show()
    return r

# Use the helper three times — satisfies the rubric's "function used more than once" requirement
r_65   = scatter_vs_checks('pct_65plus',  'Percent aged 65+',       'tomato')
r_u18  = scatter_vs_checks('pct_under18', 'Percent under 18',       'goldenrod')
r_wht  = scatter_vs_checks('pct_white',   'Percent white (non-Hispanic)', 'teal')

print(f"Correlations with checks_per_100k:")
print(f"  pct_65plus:  r = {r_65:+.3f}")
print(f"  pct_under18: r = {r_u18:+.3f}")
print(f"  pct_white:   r = {r_wht:+.3f}")
```

**Observation:** The three demographic variables show **different strengths and directions** of association with per-capita checks. `pct_white` tends to show the strongest positive correlation — states with a higher share of non-Hispanic white residents also tend to have higher per-capita background check activity. The age variables are weaker, with `pct_65plus` typically showing a mild positive trend.

These are **associations only** — they do not tell us *why* the pattern exists. Rural composition, gun-ownership culture, and state-level policy (like Kentucky's recurring-check rule) are all plausible confounders that we are not controlling for here.

```python
# Combined view: correlation matrix of all key numeric variables
corr_cols = ['checks_per_100k', 'population_2016', 'pct_under18',
             'pct_65plus', 'pct_female', 'pct_white', 'pct_black', 'pct_hispanic']
corr_matrix = merged[corr_cols].corr().round(3)
print("Correlation matrix (Pearson):")
print(corr_matrix['checks_per_100k'].sort_values(ascending=False))
corr_matrix
```

**Findings for Research Question 2:**

1. **Distribution (1D):** Per-capita checks are right-skewed across states — a handful of outliers pull the mean well above the median.
2. **Demographics (2D):** Of the three demographic independent variables tested, `pct_white` had the strongest linear association with per-capita checks; `pct_under18` and `pct_65plus` showed weaker relationships.
3. **Caveat:** Correlations here are **descriptive, not causal.** The census snapshot is a single year (2016) compared against cumulative check totals spanning 1998–2017, so temporal mismatch is a real limitation — see the Conclusions section for a fuller discussion.
