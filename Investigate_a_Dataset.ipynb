
### Data Cleaning
> **Tip**: Make sure that you keep your reader informed on the steps that you are taking in your investigation. Based on the initial exploration, the gun dataset has three issues that need attention:

1. **`month` is stored as a string** — we convert it to a proper `datetime` so we can do time-based analysis.
2. **Many columns have missing values** — columns like `other`, `permit_recheck`, `rentals_*`, and `returned_*` are sparse because the FBI added those reporting categories in later years. We fill these with `0` since a missing value here means "not yet reported," not "unknown amount."
3. **Duplicates** — we check for and drop any duplicate `(state, month)` rows.

```python
# 1. Convert 'month' to datetime and extract year for later grouping
df['month'] = pd.to_datetime(df['month'])
df['year'] = df['month'].dt.year
print("month dtype:", df['month'].dtype)
df[['month','year']].head()
```

```python
# 2. Fill NaN in numeric gun-type columns with 0
#    (missing = FBI had not yet added that category for that month)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

def fill_missing_with_zero(frame, columns):
    """Fill NaN in the given columns with 0 and return how many cells were filled."""
    before = frame[columns].isnull().sum().sum()
    frame[columns] = frame[columns].fillna(0)
    return before

filled = fill_missing_with_zero(df, numeric_cols)
print(f"Filled {filled} missing numeric cells with 0")
print(f"Remaining nulls in df: {df.isnull().sum().sum()}")
```

```python
# 3. Check for duplicate state-month rows
dup_count = df.duplicated(subset=['state','month']).sum()
print(f"Duplicate (state, month) rows: {dup_count}")

# Drop if any exist
if dup_count > 0:
    df = df.drop_duplicates(subset=['state','month']).reset_index(drop=True)
    print(f"After dedup, shape: {df.shape}")
else:
    print("No duplicates found — shape unchanged:", df.shape)
```

```python
# Final sanity check on the cleaned gun dataframe
print("Shape:", df.shape)
print("Date range:", df['month'].min(), "→", df['month'].max())
print("Unique states/territories:", df['state'].nunique())
df.info()
```

### Loading the U.S. Census Data

The census dataset is structured in **wide format** — each row is a demographic fact and each column is a U.S. state. We need to transpose it so that each row represents one state, then extract the key columns we'll use for analysis (population and demographic percentages).

```python
# Load the census data (wide format: rows=facts, columns=states)
census_raw = pd.read_csv('us_census_data.csv')

# Transpose so rows = states, columns = facts; set state names as index
census_T = census_raw.set_index('Fact').drop(columns=['Fact Note']).T
census_T.index.name = 'state'
census_T.reset_index(inplace=True)

# Select only the columns we'll use for analysis
CENSUS_COLS = {
    'Population estimates, July 1, 2016,  (V2016)': 'population_2016',
    'Persons under 18 years, percent, July 1, 2016,  (V2016)': 'pct_under18',
    'Persons 65 years and over, percent,  July 1, 2016,  (V2016)': 'pct_65plus',
    'Female persons, percent,  July 1, 2016,  (V2016)': 'pct_female',
    'White alone, not Hispanic or Latino, percent, July 1, 2016,  (V2016)': 'pct_white',
    'Black or African American alone, percent, July 1, 2016,  (V2016)': 'pct_black',
    'Hispanic or Latino, percent, July 1, 2016,  (V2016)': 'pct_hispanic',
}

census = census_T[['state'] + [c for c in CENSUS_COLS if c in census_T.columns]].copy()
census.rename(columns=CENSUS_COLS, inplace=True)

# Strip commas/percent signs and convert to numeric
def clean_numeric(series):
    """Remove commas/percent signs and cast to float."""
    return pd.to_numeric(series.astype(str).str.replace(',', '', regex=False)
                                           .str.replace('%', '', regex=False)
                                           .str.strip(), errors='coerce')

for col in census.columns[1:]:
    census[col] = clean_numeric(census[col])

print(f"Census shape: {census.shape}")
census.head()
```

### Merging FBI Gun Data with Census Data

The gun dataset has one row per **state × month**, covering multiple years. Before merging with the census data (which provides a single 2016 population snapshot), we aggregate the gun data by state — summing the total background checks and gun type counts across all months. We then perform an **inner join** on state name, which drops entries not present in both datasets (e.g., "District of Columbia" and territories absent from the census table).

```python
# Aggregate gun data by state (sum numeric columns across all months)
gun_by_state = (df.groupby('state', as_index=False)
                  .agg(total_checks=('totals', 'sum'),
                       total_handgun=('handgun', 'sum'),
                       total_long_gun=('long_gun', 'sum'),
                       total_permit=('permit', 'sum'),
                       total_multiple=('multiple', 'sum')))

# Inner join — keeps only states present in both datasets
merged = gun_by_state.merge(census, on='state', how='inner')

# Derived column: background checks per 100k residents
merged['checks_per_100k'] = (merged['total_checks'] / merged['population_2016']) * 100_000

print(f"States in gun data:    {gun_by_state['state'].nunique()}")
print(f"States in census data: {census['state'].nunique()}")
print(f"States after merge:    {merged['state'].nunique()}")
print(f"\nMerged dataframe shape: {merged.shape}")
merged.head()
```

```python
# Verify no nulls in key columns after merge
print("Null counts in merged dataset:")
merged[['total_checks', 'population_2016', 'checks_per_100k']].isnull().sum()
```
