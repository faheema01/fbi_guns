## Introduction

### Dataset Description

This analysis uses two datasets:

**1. FBI NICS Gun Background Check Data (`gun_data.csv`)**  
The National Instant Criminal Background Check System (NICS) is used by Federal Firearms Licensees (FFLs) to determine whether a prospective buyer is eligible to buy firearms. The dataset contains monthly background check counts by U.S. state from **November 1998 through 2017**.

Key columns include:

| Column | Description |
|---|---|
| `month` | Year-month of the checks (YYYY-MM) |
| `state` | U.S. state or territory |
| `permit` | Permit checks |
| `handgun` | Handgun background checks |
| `long_gun` | Long gun (rifle/shotgun) background checks |
| `multiple` | Multiple firearm checks |
| `totals` | Total background checks across all categories |
| `prepawn_handgun/long_gun` | Pre-pawn checks |
| `redemption_handgun/long_gun` | Redemption checks |
| `private_sale_handgun/long_gun` | Private sale checks |

**2. U.S. Census Data (`us_census_data.csv`)**  
Contains demographic and population estimates by state (circa 2016). Relevant columns include state population, age distributions, racial demographics, income, and education levels. This dataset enables us to normalize gun check counts by population.

The two datasets are linked by **state name**, allowing per-capita analysis.

---

### Research Questions

This analysis investigates the following questions:

1. **How have total gun background checks trended over time in the U.S.?**  
   *(Dependent variable: `totals`; Independent variable: `month`/`year`)*

2. **Which states have the highest gun background check rates, and how does this compare when normalized by population?**  
   *(Dependent: `totals`; Independent: `state`, Census `population`)*

3. **What is the breakdown between handgun and long gun checks by state — and has that ratio changed over time?**  
   *(Dependent: `handgun` vs `long_gun`; Independent: `state`, `year`)*

4. **Do states with higher populations show proportionally higher background check volumes?**  
   *(Dependent: per-capita checks; Independent: Census population)*
